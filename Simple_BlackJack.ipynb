```python
import random
```

```python
suits = ('Hearts','Spades','Diamonds','Clubs')
ranks = ('Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King')
```

```python
values = {'Ace':11,'Two':2,'Three':3,'Four':4,
                'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10}
```

```python
playing = True
```

```python
class Card():
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
    def __str__(self):
        return "{} of {}".format(self.rank,self.suit)
```

```python
class Deck():
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        string = ''
        for card in self.deck:
            string = string + '\n' + str(card)
        return 'The deck contains: \n' + string
    
    def shuffle(self):
        return random.shuffle(self.deck)
    
    def deal(self):
        return self.deck.pop()
        
        
```

```python
class Hand():
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
    
    def add_card(self,card):
        self.cards.append(card)
        self.value = self.value + values[card.rank]
        if card.rank == 'Ace':
            self.aces = self.aces + 1
            
    def adjust_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
    
```

```python
class Chips():
    def __init__(self):
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    while True:
        try:
            chips.bet = int(input('input(How much do you want to bet?'))
        except ValueError:
            print('Sorry! Bet should be an integer!!')
            print('Try Again!')
            continue
        else:
            if chips.bet  > chips.total:
                print('Bet should not exceed your total amount:',chips.total)
                print('Try Again!')
                continue
            else:
                break
```

```python
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_ace()
    
```

```python
def hit_or_stand(deck,hand):
    global playing
    while True:
        decision = input('Would you like to hit or stand (h/s): ')
        if decision[0].lower() == 'h':
            hit(deck,hand)
        elif decision[0].lower() == 's':
            print('Player stands and the Dealer is dealing')
            playing = False 
        else:
            print('Please enter again whether you want to hit or stand (h/s)')
            continue
        break   
```

```python
def show_some(player,dealer):
    print("\nDealer's cards:")
    print(' <card hidden>')
    print(" ",dealer.cards[1])
    print("\nPlayer's cards:")
    print(' ',*player.cards,sep = "\n")
def show_all(player,dealer):
    print("\nDealer's cards:\n")
    print(" ",*dealer.cards,sep = '\n')
    print("\nPlayer's cards:\n")
    print(' ',*player.cards,sep = "\n")
```

```python
def player_busts(player,dealer,chips):
    print("Player busts!")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("Player wins!")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("Dealer busts!")
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print("Dealer wins!")
    chips.lose_bet()
    
def push(player,dealer):
    print("Dealer and Player tie! It's a push.")
```
