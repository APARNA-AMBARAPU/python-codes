```python
def is_tictactoe(pos):
    pos['1'].sort()
    pos['2'].sort()
    l1 = pos['1']
    l2 = pos['2']
    l = []
    for i in range(9):
        if i in l1:
            l.append('o')
        elif i in l2:
            l.append('x')
        else:
            l.append(' ')
    mark = ['o','x']
    for m in mark:
        if l[0] == m and l[1] == m and l[2] == m:
            return m, True
        elif l[3] == m and l[4] == m and l[5] == m:
            return m, True
        elif l[6] == m and l[7] == m and l[8] == m:
            return m, True
        elif l[0] == m and l[3] == m and l[6] == m:
            return m, True
        elif l[1] == m and l[4] == m and l[7] == m:
            return m, True
        elif l[6] == m and l[7] == m and l[8] == m:
            return m, True
        elif l[0] == m and l[4] == m and l[8] == m:
            return m, True
        elif l[2] == m and l[4] == m and l[6] == m:
            return m, True
        else:
            if m == 'o':
                continue
            else:
                if l.count(' ') == 0:
                    win_bool = 'd'
                    return None,win_bool
                else:
                    return None,None
            
        
def print_tictactoe(pos):
    pos['1'].sort()
    pos['2'].sort()
    l1 = pos['1']
    l2 = pos['2']
    l = []
    for i in range(9):
        if i in l1:
            l.append('o')
        elif i in l2:
            l.append('x')
        else:
            l.append(' ')
            
    print(" {} | {} | {} ".format(l[0],l[1],l[2]))
    print("____________")
    print(" {} | {} | {} ".format(l[3],l[4],l[5]))
    print("____________")
    print(" {} | {} | {} ".format(l[6],l[7],l[8]))
    print("____________")
```

```python
count = 0
pos = {'1':[],'2':[]}
while count < 11 :
    print('Enter the position of your "o":')
    p1 = input()
    pos['1'].append(int(p1))
    
    if count != 8:
        print('Enter the position of your "x":')
        p2 = input()
        pos['2'].append(int(p2))

    count = count + 2
    symbol,won = is_tictactoe(pos)
    print_tictactoe(pos)
    if won and symbol == 'o':
        print("Player 1 won the match")
        break
    if won and symbol == 'x':
        print('Player 2 won the match')
        break
    if won == 'd':
        print('It is a draw')
        break
    if won == None and symbol == None:
        continue
```
